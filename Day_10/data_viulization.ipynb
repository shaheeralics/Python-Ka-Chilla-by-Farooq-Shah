#   plotting   in  python

```python
#import libraries
import  pandas  as  pd
import  numpy  as  np
import  matplotlib.pyplot  as  plt
import  seaborn  as  sns
```

```python
#load   datasets
kashti=sns.load_dataset('titanic')
kashti
```

```python
kashti.isnull().sum()
```

```python
#barplotting
sns.countplot(x="survived",hue="sex",data=kashti)
```

```python
#scatterplotting
plt.scatter(data=kashti,x="age",y="fare")
```

```python
#further addition    inplot  through seaborn library
sns.scatterplot(data=kashti,x="age",y="fare",hue="pclass")
```

```python
#barplotting
sns.barplot(data=kashti,x="sex",y="fare",hue="pclass")
```

```python
#pairplotting
sns.pairplot(kashti)
```

```python
#histplotting
kashti.hist(bins=6,figsize=(9,7),grid=False)
```

```python
#boxplotting
sns.boxplot(data=kashti,x="sex",y="fare",hue="embarked")
```

```python
#boxenplot
sns.boxenplot(data=kashti,x="sex",y="fare",hue="pclass")
```

```python
# Create a FacetGrid
g = sns.FacetGrid(kashti, col="pclass", hue="survived", margin_titles=True, palette={1:"seagreen",0:"gray"})

# Map a scatter plot to the grid
g.map(plt.scatter, "age", "fare",edgecolor="w").add_legend()
```

```python
#stripplotting
sns.stripplot(data=kashti,x="sex",y="fare",hue="embarked")
```

```python
import pandas as pd

# Load the Titanic dataset (assuming it is in CSV format)
titanic_data = kashti

# Select the desired columns for correlation
selected_columns = ['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare']
selected_data = titanic_data[selected_columns]

# Compute the correlation matrix
correlation_matrix = selected_data.corr()

# Display the correlation matrix
print("Correlation matrix:")
print(correlation_matrix)
```

```python
# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=.5)


```

```python
#voilinplotting
sns.violinplot(data=kashti,x="sex",y="fare",hue="class")
```
