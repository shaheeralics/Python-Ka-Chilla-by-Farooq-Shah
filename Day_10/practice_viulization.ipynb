```python
#importing  libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading    datasets
pholl=sns.load_dataset('iris')
```

```python
pholl
```

```python
pholl.sample(100)
```

```python
pholl.isnull()
```

```python
# Create bar plots for each feature by species
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for feature in features:
    plt.figure(figsize=(8, 6))
    sns.barplot(x='species', y=feature, data=pholl, ci=None,hue="species")
    
```

```python
# Create scatter plots for each pair of features by species
sns.pairplot(pholl, hue='species', markers=["o", "s", "D"])
```

```python
# Create histograms for each feature by species

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for feature in features:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=pholl, x=feature, hue='species', multiple='stack', kde=True)
```

```python
# Create box plots for each feature by species

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for feature in features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='species', y=feature, data=pholl)
    
```

```python
# Create boxen plots for each feature by species
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for feature in features:
    plt.figure(figsize=(8, 6))
    sns.boxenplot(x='species', y=feature, data=pholl)
```

```python

# Create a FacetGrid to visualize pairwise relationships by species
g = sns.FacetGrid(pholl, col="species", height=4, aspect=0.7)
g.map(sns.scatterplot, "sepal_length", "sepal_width")
g.add_legend()
```

```python
# Create strip plots for each feature by species
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for feature in features:
    plt.figure(figsize=(8, 6))
    sns.stripplot(x='species', y=feature, data=pholl, jitter=True)
```

```python
# Select only numeric columns for correlation matrix
numeric_iris = pholl.select_dtypes(include=['float64'])

# Calculate the correlation matrix
corr_matrix = numeric_iris.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Iris Dataset')
```
