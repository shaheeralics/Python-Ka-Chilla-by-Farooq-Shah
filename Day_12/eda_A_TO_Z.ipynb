#   exploratry  data    analysis

Exploratory Data Analysis (EDA) is a crucial step in the data analysis process, where the primary goal is to understand the underlying structure and patterns within a dataset. EDA involves summarizing the main characteristics of the data, often with visual methods, before formal modeling or hypothesis testing. This process helps in identifying anomalies, discovering patterns, and forming hypotheses that can guide further analysis.

```python
#viulize    and explore
sns.countplot(data=df,x="survived")
```

```python
sns.barplot(data=df,x="survived",y="age",hue="who")
```

```python
sns.countplot(data=df,x="pclass")
```

```python
#data   distribution
sns.histplot(data=df,x='age',bins=50)
```

```python
df['age'].hist(bins=30,color='darkred',alpha=0.7)
```

```python
sns.boxplot(data=df,y='age',x='survived',hue='who')
```

```python
#scatterplotting
sns.scatterplot(data=df,x='age',y='fare',hue='pclass')
plt.ylim(0,300)
```

```python
sns.pairplot(df,hue='survived')
```

```python
# Group by 'pclass' and 'sex', calculate mean of 'survived', and unstack to reshape the DataFrame
grouped= df.groupby(['pclass', 'who'])['fare'].mean().unstack()

# Plot the stacked bar chart
grouped.plot(kind='bar', stacked=True)
```

```python
sns.swarmplot(data=df,x='survived',y='age',hue='who')
```

```python
#catplotting
sns.catplot(x='pclass', y='fare', hue='embarked', kind='point', data=df)
```

```python
#relation   between two numeric variables
sns.lmplot(data=df,x='fare',y='age')
```

```python


# Filter the dataset to show only ages less than 30
df2= df[df['fare'] < 300]

# Create an lmplot
sns.lmplot(x='age', y='fare', hue='pclass', data=df2)
```

```python
#grouping
df.columns
```

```python
df['age_groups']=pd.cut(df['age'],[0,10,18,30,50,80],labels=['Kids','Child','Youth','Middle aged','Old'])
```

```python
df.head()
```

```python
df.groupby("age_groups")["fare"].mean().plot(kind='bar')
```

```python
sns.pairplot(df,hue="survived",vars=['age','fare','pclass'])
```
