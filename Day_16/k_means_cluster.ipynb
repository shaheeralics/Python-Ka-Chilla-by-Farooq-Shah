```python
# import libraries
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
df.head()
df1=df
df1=pd.get_dummies(df1,columns=['sex'])
```

```python
# model 

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler # for data normalization


# importing the data in features
x = df[['age', 'fare']]
x.age.fillna(x.age.mean(), inplace=True)

# scaling the data

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# perform clustering

model = KMeans(n_clusters = 2)
model.fit(x_scaled)

# assign cluster label to the data

df['cluster'] = model.labels_
```

```python
plt.scatter(df['age'],df['fare'],c=df['cluster'], cmap= 'rainbow')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['age'],df['fare'],c=df['cluster'], cmap= 'rainbow')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1], color = 'black')
plt.ylabel('Fare')
plt.xlabel('Age')
plt.title('Idea Of 3 Types of Age Clusters \nwith the Fare of the Passangers\n in the Titanic')
plt.show()
```
