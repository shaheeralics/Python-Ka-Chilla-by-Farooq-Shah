```python
import  pandas  as  pd
import  numpy  as  np
import  matplotlib.pyplot  as  plt
import  seaborn  as  sns
import  plotly  as  px
```

```python
import matplotlib.pyplot as plt

# Data for the last 10 years
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
population = [185.1, 189.5, 193.9, 197.7, 202.3, 206.7, 210.8, 215.3, 220.9, 225.2]  # in millions

# Create the plot
plt.figure(figsize=(9, 4))
plt.plot(years, population, marker='o', linestyle='-', color='b')

# Add title and labels
plt.title('Population of Pakistan (2014-2023)')
plt.xlabel('Year')
plt.ylabel('Population (millions)')

# Display the plot
plt.grid(True)
plt.show()
```
